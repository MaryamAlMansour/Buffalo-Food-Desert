### Step 3: Define the Data Model
#### 3.1 Conceptual Data Model
Map out the conceptual data model and explain why you chose that model

Star Schema will be the 

Fact Table (Food Desert):
Fact table contains the keys to associated dimension tables. it contains all the possible measurment and facts to locate a food desert area in Buffalo, based on the number of aggregated stores within that area. Each store will be tagged with license number and zipcode that locate its place. 

Fact Table..
Food Desert Column Names:
1. license_number: Identifies what type of store is this. 
2. city, county, state, zip_code: All those entities complete a particular location of the store. 
3. dba_name: the commerical name of the store. “Doing business as” (DBA) name for the licensed entity
4. square_footage: the aproximate space of the store.   

Dimension Tables..
The dimension tables contain descriptive information about the numerical values in the fact table.

License Dimension Column Names: 
1. License number: which identify the licene of a particular store
2. operation_type: The type of operation the licensed entity conducts, for this dataset, it is all stores. 
3. establishment_type: The establishment type for the licensed entity at this location; search for NYSDAM_RetailFoodStoresEstablishmentTypeCodes.pdf
4. entity_name: The name of the licensed entity (name of the store proffissonally)

Store Location Dimension Column Names: 
1. Location: contain address, city, state and longitude + latitude of a speicifc store in Buffalo and its near neighbours. 
2. zip_code: which identifies each city in buffalo, There are around 20 zipcodes [4201, 14202, 14203, 14204, 14206, 14207, 14208, 14209, 14210, 14211..] will be used to locate a particular store. 
3. lanitude, latitude: used for geolocation. Please use this map for an interactive location of the retail store. https://data.ny.gov/Economic-Development/Retail-Food-Stores-Map/p2dn-xhaw

Vacant Land Dimension Column Names:
1. property_type: Property Type Classification Codes describe the primary use of each parcel of real property, in our case we filtered it only to vacant land which is 311. 
2. property_class_description: A description of the primary use of each parcel of real property. Residentila, industrial, commerical, park lot, gas station..etc. 
3. owner1: The owner of a parcel of real propertyt, sometimes an indivisual, sometimes an orgnization name.
4. city,address,state: Gives the address of this vacant land. 

Vland location Dimension Column Names: 
1. address, zip_code: will give the specific district and city of this vacant land
2. longitude, latitude: used for geolocation purposes.
